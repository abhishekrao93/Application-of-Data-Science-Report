```python
import pandas as pd

raw_data=pd.read_csv("C:\\Users\\unbea\\Downloads\\45542791_ITEC876_Ass2\\yelp.csv")
yelp_data=pd.DataFrame(data=raw_data, columns=['stars','text'])
yelp_data
```

```python
# Loading data from csv file and separating each text into positive and negative based on the stars criteria 
x=yelp_data.loc[:,'stars']
y=yelp_data.loc[:,'text']
counter_positive=1
counter_negative=1
counter=0
# Storing each review in text file based on the stars
for i in x:
# 4 and 5 stars are treated as positive feedback
    if i == 4 or i == 5: 
            f=open("C:\\Users\\unbea\\Downloads\\45542791_ITEC876_Ass2\\Yelp_Positive\\Positive_"+str(counter_positive)+".txt","w+",encoding="utf-8")
            f.write(y[counter])
            f.close()
            counter_positive=counter_positive +1
# 1 ,2 and 3 stars are treated as negative feedback
    else: 
            f=open("C:\\Users\\unbea\\Downloads\\45542791_ITEC876_Ass2\\Yelp_Negative\\Negative_"+str(counter_negative)+".txt","w+",encoding="utf-8")
            f.write(y[counter])
            f.close()
            counter_negative=counter_negative +1
    counter=counter + 1
```
