```python
# Storing all tweets in a list
avengers_list=[]

for line in open('Avengers_2.json', 'r'):
    avengers_list.append(json.loads(line))
   
```

```python
#Function made so that we can separate tweets and store it in list and extract links and hashtags
emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""
string_re = [
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
 
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
    r'(?:[\w_]+)', # other words
    r'(?:\S)' # anything else
]
    
tokens_re = re.compile(r'('+'|'.join(string_re)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
 
def searching(s):
    return tokens_re.findall(s)
 
def cleaning(s, lowercase=False):
    tokens = searching(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens
```

```python
#Converting Raw tweet into simple text by removing hasthags and links
list_1=[]
store_tweets=[]
for line in avengers_list:
        s=""
        list_1=cleaning((line['full_text']))
        print(list_1)
        for i in list_1:
            if (i.startswith('#') or i.startswith('https') or i.startswith('@') or i.startswith('RT') or i.startswith(':')):
                i
            else:
                s=s+" "+ i
        print(s)
        store_tweets.append(s)
```
